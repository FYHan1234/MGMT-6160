```python
#Plug another data set in these codes to get results for New York 2017, New York 2016, Alabama 2016 results.
import pandas as pd
col=["tract_to_msamd_income",
"population",
"minority_population",
"number_of_owner_occupied_units",
"number_of_1_to_4_family_units",
"loan_amount_000s",
"hud_median_family_income",
"applicant_income_000s",
"msamd_name",
"loan_type_name",
"loan_purpose_name",
"lien_status_name",
"hoepa_status_name",
"applicant_sex_name",
"applicant_race_name_1",
"applicant_ethnicity_name",
"agency_name",
"action_taken_name"]
AL2017=pd.read_csv('./hmda_lar.csv',usecols=col)
AL2017.head()
```

### Some visualizations### 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='applicant_race_name_1', data=AL2017) 
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='action_taken_name', data=AL2017) 
```

### Missing values### 

```python
AL2017._get_numeric_data().columns
```

```python
AL2017[AL2017._get_numeric_data().columns].isnull().sum()
```

```python
AL2017['tract_to_msamd_income']=AL2017['tract_to_msamd_income'].fillna(AL2017['tract_to_msamd_income'].mean())
AL2017['population']=AL2017['population'].fillna(AL2017['population'].mean())
AL2017['minority_population']=AL2017['minority_population'].fillna(AL2017['minority_population'].mean())
AL2017['number_of_owner_occupied_units']=AL2017['number_of_owner_occupied_units'].fillna(AL2017['number_of_owner_occupied_units'].mean())
AL2017['number_of_1_to_4_family_units']=AL2017['number_of_1_to_4_family_units'].fillna(AL2017['number_of_1_to_4_family_units'].mean())
```

```python
AL2017['loan_amount_000s']=AL2017['loan_amount_000s'].fillna(AL2017['loan_amount_000s'].mean())
AL2017['hud_median_family_income']=AL2017['hud_median_family_income'].fillna(AL2017['hud_median_family_income'].mean())
AL2017['applicant_income_000s']=AL2017['applicant_income_000s'].fillna(AL2017['applicant_income_000s'].mean())
```

```python
#Now check other columns to see if there are NA values
AL2017.isnull().sum()
```

```python
#Remove 8518 rows that don't contain metro areas information
AL2017=AL2017.dropna()
```

```python
#Check NA values again
AL2017.isnull().sum()
AL2017
```

```python
#Assign 0 to denied, and 1 to approved
AL2017.loc[AL2017['action_taken_name']=='Application denied by financial institution','action_taken_name']=0
AL2017.loc[AL2017['action_taken_name']=='Application approved but not accepted','action_taken_name']=1
```

```python
#Seperate the target before create dummy variables
y=AL2017['action_taken_name']
AL2017.drop(['action_taken_name'],axis=1)
```

```python
#Dummy variables
AL2017=pd.get_dummies(AL2017)
```

```python
#Concatenate AL2017 with y again
AL2017=pd.concat([AL2017,y],axis=1)
```

```python
AL2017
```

### Data partitioning###

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(AL2017, test_size=0.3)
```

```python
train.head()
```

```python
test.head()
```

### Modeling###

```python
#Logistic regression from ScikitLearn Package 
from sklearn.ensemble import RandomForestClassifier
X_train=train.iloc[:,:47]
X_train.head()
```

```python
y_train=train.iloc[:,47]
y_train.head()
```

```python
clf=RandomForestClassifier(n_estimators=500, max_depth=4,random_state=0)
clf.fit(X_train, y_train)
```

```python
pred_train_AL2017=clf.predict(X_train)
pred_train_AL2017
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
accuracy_score(y_train,pred_train_AL2017) 
```

### Testing###

```python
X_test=test.iloc[:,:47]
X_test.head()
```

```python
y_test=test.iloc[:,47]
y_test.head()
```

```python
pred_test_AL2017=clf.predict(X_test)
pred_test_AL2017
```

```python
accuracy_score(y_test,pred_test_AL2017) 
```

### Final analysis###

```python
#Now the final analysis
#Approval rate in the test set (using predictions)
app1=np.count_nonzero(pred_test_AL2017)/len(y_test) 
app1
```
